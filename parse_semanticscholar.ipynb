```python
import wget
import time
import pandas as pd
import os
import json
```

```python
# Load current anthology 
df = pd.read_csv("data/anthology_conferences.csv", sep="|", keep_default_na=False)
df[:2]
```

```python
# Download semantic scholar paper information by doi
download = 0
start = time.time()

no_doi = 0
not_found = 0

for i, row in df.iterrows():
    if row["doi"] != "":
        try:
            if not os.path.exists("data/semantic_scholar/papers/" + row["doi"].split("/")[-1] + ".json"):
                filename = wget.download("https://api.semanticscholar.org/v1/paper/" + row["doi"], out="data/semantic_scholar/papers/" + row["doi"].split("/")[-1] + ".json")
                #print(filename)
                download += 1
        except:
            print("Could not download doi:", row["doi"])
            not_found += 1
    else:
        print("Entry with no doi")
        #print(row)
        no_doi += 1
    t = time.time()
    if download >= 100 and t-start < 290:
        print("Waiting with time " + str(t-start) + "...")
        while time.time() - start < 300:
            continue
        print("Continued with time " + str(time.time()-start) + ".")
        download = 0
        start = time.time()
```

```python
print("Number of entries with scholar entries:", len(df[df["semantic_scholar"] != ""]))
print("Number of papers with doi but without scholar entries:", not_found)
print("Number of papers with no doi and without scholar entries:", no_doi)
```

```python
# Add scholar keywords to dataframe
df["semantic_scholar_keywords"] = ""

id2topic = dict()

for i, row in df.iterrows():
    if row["semantic_scholar"] != "":
        with open("data/semantic_scholar/papers/" + row["semantic_scholar"]) as jf:
            data = json.load(jf)
        topics = []
        for topic in data["topics"]:
            topics.append(topic["topicId"])
            if topic["topicId"] not in id2topic:
                id2topic[topic["topicId"]] = topic["topic"]
        df.at[i, "semantic_scholar_keywords"] = topics
    else:
        df.at[i, "semantic_scholar_keywords"] = []
```
